# Cluster product descriptions into similar groups by using NLP techinque

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Omkar\Downloads\Myntra dataset\Myntra fashion\Myntra Fasion Clothing.csv")
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

documents = data['Description'].values.astype("U")

vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
```

```python
k = 20
model = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=1)
model.fit(features)
data['cluster'] = model.labels_
data.head()
```

```python
# output the result to a text file.

clusters = data.groupby('cluster')    

for cluster in clusters.groups:
    f = open('cluster'+str(cluster)+ '.csv', 'w') # create csv file
    data = clusters.get_group(cluster)[['Individual_category','Description']] # get title and overview columns
    f.write(data.to_csv(index_label='id')) # set index to id
    f.close()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms  = vectorizer.get_feature_names_out()

for i in range(k):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :10]: #print out 10 feature terms of each cluster
        print (' %s' % terms[j])
    print('------------')
```
